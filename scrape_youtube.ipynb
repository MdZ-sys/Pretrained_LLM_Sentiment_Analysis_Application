```
def get_video_details(youtube, video_id):
    response = youtube.videos().list(
        part='snippet',
        id=video_id
    ).execute()

    video_details = response['items'][0]['snippet']
    channel_name = video_details['channelTitle']
    published_date = datetime.strptime(video_details['publishedAt'], '%Y-%m-%dT%H:%M:%S%z').date()
    return channel_name, published_date

def get_video_comments(youtube, video_ids):
    comments_data = []

    video_request = youtube.commentThreads().list(
        part='snippet',
        videoId=','.join(video_ids),
        maxResults=50  # You can increase this
    )
    video_response = video_request.execute()

    video_details = {}

    for item in video_response['items']:
        video_id = item['snippet']['videoId']
        comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
        username = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
        published_date = datetime.strptime(item['snippet']['topLevelComment']['snippet']['publishedAt'], '%Y-%m-%dT%H:%M:%S%z').date()

        channel_name, _ = video_details.get(video_id, ("", None))

        comments_data.append({
            'Date': published_date,
            'Media Source': 'YouTube Comments',
            'Domain Name': channel_name,
            'URL': f'https://www.youtube.com/watch?v={video_id}',
            'Likes': 0,
            'Comments': 0,
            'Shares': 0,
            'Is Reply': False,
            'Commenter Name': username,
            'Comment Text': comment
        })

    return comments_data

def scrap_youtube_comments(query, num_of_videos):
    youtube = build('youtube', 'v3', developerKey=API_KEY)

    video_data = search_videos(youtube, query, num_of_videos)
    video_ids = [video['videoId'] for video in video_data]

    # Fetch video details for all video_ids
    video_details = {video_id: get_video_details(youtube, video_id) for video_id in video_ids}

    comments_data = get_video_comments(youtube, video_ids)
    return comments_data

def search_videos(youtube, query, num_of_videos):
    request = youtube.search().list(
        part='snippet',
        q=query,
        type='video',
        maxResults=num_of_videos
    )
    response = request.execute()
    return [{'url': f'https://www.youtube.com/watch?v={item["id"]["videoId"]}', 'videoId': item['id']['videoId']} for item in response['items']]
```
